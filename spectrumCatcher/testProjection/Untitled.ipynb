```python
FOVimage = np.copy(img[500-480:670-480, 180-70:230-70, :])
plt.imshow(FOVimage)
FOVimage.shape
```

```python
fovAzimuth = 10/180*np.pi
fovPitch   = 60/180*np.pi
fov_dAzi_2 = 1/180*np.pi
fov_dPit_2 = 5/180*np.pi

FoV = np.zeros((170, 50, 5))
FoV[:,:,2:] =  FOVimage
FoV[:,:,0] = np.linspace(fovAzimuth+fov_dAzi_2, fovAzimuth-fov_dAzi_2, num=50).reshape(1, 50)
FoV[:,:,1] = np.linspace(fovPitch-fov_dPit_2  , fovPitch+fov_dPit_2  , num=170).reshape(170,1)

# plt.imshow(FoV[:,:,0])
# plt.show()
# plt.imshow(FoV[:,:,1])
# plt.show()
```

```python
tmpx = int(400/np.tan(fovPitch) * np.cos(fovAzimuth))
tmpy = int(400/np.tan(fovPitch) * np.sin(fovAzimuth))
print(tmpx, tmpy)

def indexNearest(givenAzimuth, givenPitch, heightPx = 400, center_adjustX= 500, center_adjustY = 500):
    tmpx = (heightPx/np.tan(givenPitch) * np.cos(givenAzimuth)) + center_adjustX
    tmpy = (heightPx/np.tan(givenPitch) * np.sin(givenAzimuth)) + center_adjustY
    return int(round(tmpx)), int(round(tmpy))

point0 = indexNearest(FoV[ 0, 0, 0], FoV[ 0, 0, 1])
point1 = indexNearest(FoV[ 0,-1, 0], FoV[ 0,-1, 1])
point2 = indexNearest(FoV[-1,-1, 0], FoV[-1,-1, 1])
point3 = indexNearest(FoV[-1, 0, 0], FoV[-1, 0, 1])
print(point0, point1, point2, point3)
plt.figure(figsize= (15,15))
plt.imshow(a[:,:,1])
plt.scatter(tmpx+500, tmpy+500)
plt.scatter(point0[0], point0[1], c='r', s = 2)
plt.scatter(point1[0], point1[1], c='r', s = 2)
plt.scatter(point2[0], point2[1], c='r', s = 2)
plt.scatter(point3[0], point3[1], c='r', s = 2)
```
